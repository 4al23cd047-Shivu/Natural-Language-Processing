```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('vader_lexicon')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk import pos_tag
```

```python
# text='''I really love learning Natural Language Processing . It is very interesting'''
text="I do not want to attend tommarows event . Its totally time Waste"
```

```python
print("Raw Text:\n",text)
```

```python
token=word_tokenize(text)
print("Tokens:",token)
```

```python
stop_words=set(stopwords.words('english'))
```

```python
print("Stop words:",stop_words)
```

```python
filtered_tokens=[]
for word in token:
  if word.lower() not in stop_words:
    filtered_tokens.append(word)
```

```python
print("After stop word removal:",filtered_tokens)
```

```python
stemmer=PorterStemmer()
stemmed_words=[]
for word in filtered_tokens:
  stemmed_words.append(stemmer.stem(word))
print("After stemming:",stemmed_words)
```

```python
lemmatizer=WordNetLemmatizer()
lemmatized_words=[]
for word in filtered_tokens:
  lemmatized_words.append(lemmatizer.lemmatize(word))
print("After lemmatization:",lemmatized_words)
```

```python
pos_tags=pos_tag(filtered_tokens)
print("Pos taggs:",pos_tags)
```

```python
grammer="NP:{<DT>?<JJ>*<NN>}"
```

```python
parser_data=nltk.RegexpParser(grammer)
```

```python
tree=parser_data.parse(pos_tags)
print(tree)
```

```python
s=SentimentIntensityAnalyzer()
sentiment=s.polarity_scores(text)
print("Sentiment Analysis:",sentiment)
```

```python
#Compound score ranges between -1 to 1
#>0.05  Positive
#<-0.05  Negative
#-0.05 to 0.05 neutral
```
